```
#make the column car_name the index
#https://stackoverflow.com/questions/38542419/could-pandas-use-column-as-index
auto_dataset.set_index('car_name', inplace=True)

display(auto_dataset)
first_rows_data = auto_dataset.iloc[:3,:]
```

```
auto_dataset.iloc[:,-1]
```

```
#Split the dataset in training (70%) and test (30%) sets.

#let's make y the label vector
X = auto_dataset.iloc[:,:-1]
#the rest
y = auto_dataset.iloc[:,-1]

from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.3, train_size=.7)
```

```
#use sklearn to train and test a linear regression model.
from sklearn import datasets, linear_model
from sklearn.metrics import accuracy_score

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(xTrain, yTrain)

yPred = regr.predict(xTest)
```

```
# The coefficients

print('intercept found: ' ,regr.intercept_)
print('slope found (for each column):', regr.coef_)
```

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(yTest, yPred))
print('Mean Squared Error:', metrics.mean_squared_error(yTest, yPred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(yTest, yPred)))
```

4. Double the number of columns in the dataframe according to the following scheme: 
  * for each feature, form a new feature whose values are the squares of the values of the original feature and add it to the dataframe.
  * Apply linear regression again and calculate the mean absolute error.

```
concrete_dataset = pd.read_excel('Concrete_Data.xls')
print('Dataset:')
concrete_dataset
```

```
y = concrete_dataset.iloc[:,-1].values
X = concrete_dataset.iloc[:,:-1].values
```

```
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, train_size=.7)
```

```
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x_train, y_train)

y_pred = regr.predict(x_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
"""
Double the number of columns in the dataframe according to the following scheme:
for each feature, form a new feature whose values are the squares of the values of the original feature and add it to the dataframe.
Apply linear regression again and calculate the mean absolute error.

So, I will assume in here feature = columns?
"""


headers_list = []
for i in (concrete_dataset):
  headers_list.append(i)

print(headers_list)
```

```
y_mod = concrete_dataset.iloc[:,-1].values
X_mod = concrete_dataset.iloc[:,:-1].values
```

```
x_train_mod, x_test_mod, y_train_mod, y_test_mod = train_test_split(X_mod, y_mod, test_size = 0.3, train_size=.7)
```

```
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x_train_mod, y_train_mod)

y_pred_mod= regr.predict(x_test_mod)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test_mod, y_pred_mod))
print('Mean Squared Error:', metrics.mean_squared_error(y_test_mod, y_pred_mod))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test_mod, y_pred_mod)))
```
